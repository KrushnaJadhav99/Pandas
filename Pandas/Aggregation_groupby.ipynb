```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('SearchTrendData.csv')
```

```python
df.head()
```

```python
t_data = df[-10:]
t_data = t_data.reset_index(drop = True)
t_data
```

```python
moving_average = t_data.rolling(window=3, min_periods=1)
moving_average.aggregate(np.mean)
```

```python
moving_average = t_data.rolling(window=3, min_periods=1)
moving_average['DataScience'].aggregate(np.mean)
```

```python
moving_average = t_data.rolling(window=3, min_periods=1)
print(moving_average['DataScience'].aggregate(np.max))
print(moving_average.aggregate(np.max))
```

```python
moving_average = t_data.rolling(window=3, min_periods=1)
moving_average.aggregate(np.std)
```

```python
moving_average.aggregate(np.var)
```

```python
moving_average = t_data.rolling(window = 3, min_periods=1)
moving_average[['DataScience','MachineLearning']].aggregate([np.min, np.max, np.mean])
```
